```python
from pyspark.sql import functions as F
BASE = "/Workspace/Users/huangyue1752@gmail.com"

def read_csv(name):
    path=f"{BASE}/{name}"
    df=(spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(path))
    return df

df_client=read_csv("sample_client_source.csv")
df_product=read_csv("sample_client_product_events_source.csv")
df_channel=read_csv("sample_banking_channel_activities_source.csv")
df_consent=read_csv("sample_client_marketing_pref_source.csv")

#display(df_consent)


BRONZE=f"{BASE}/bronze"

def write_bronze(df,name):
    (df.withColumn("ingest_ts",F.current_timestamp())
     .write
     .format("delta")
     .mode("append")
     .saveAsTable(f"bronze_{name}")
    )

write_bronze(df_channel,"channel")

spark.table("bronze_channel").display()


```
