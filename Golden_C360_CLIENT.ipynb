```python
# Gold from Silver (SQL-only, run in PySpark notebook)

# Silver tables
SILVER_CLIENT = "silver_client"

# 1) C360_CLIENT (client_name hash + tenure months)
spark.sql(f"""
CREATE OR REPLACE TABLE c360_client AS
SELECT
client_id,
sha2(coalesce(cast(client_name as string), ''), 256) AS client_name,
cast(floor(months_between(current_date(), client_service_start_date)) as int) AS client_tenure_mth,
current_date() AS proc_dt
FROM {SILVER_CLIENT}
""")

spark.sql(f"select * from c360_client").display()
```
