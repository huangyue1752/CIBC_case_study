```python
# Gold from Silver (SQL-only, run in PySpark notebook)

# Silver tables
SILVER_CLIENT = "silver_client"
SILVER_PRODUCT = "silver_product"
SILVER_CHANNEL = "silver_channel"
SILVER_MKT_PREF = "silver_consent" #SCD2

# 2) C360_CLIENT_PREFERENCES (latest consent status per client)
# note：Here is_current = true；consent_update_field: 'Opt-in'/'Opt-out'
spark.sql(f"""
CREATE OR REPLACE TABLE c360_client_preferences AS
SELECT
client_id,
consent_source,
CASE
WHEN lower(consent_update_field) = 'opt-in' THEN 1
WHEN lower(consent_update_field) = 'opt-out' THEN 0
ELSE NULL
END AS latest_consent_status,
current_date() AS proc_date
FROM {SILVER_MKT_PREF}
WHERE is_current = true
""")

# Quick checks
spark.sql("SELECT * FROM c360_client_preferences").display()
```
